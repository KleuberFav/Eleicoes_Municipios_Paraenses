```python
# Informações do dataset

df.info()
```

```python
# Transformando a variável despesa em tipo float

df['VR_DESPESA_CONTRATADA'] = df['VR_DESPESA_CONTRATADA'].astype(str)
df['VR_DESPESA_CONTRATADA'] = df['VR_DESPESA_CONTRATADA'].str.replace(',', '.')
df['VR_DESPESA_CONTRATADA'] = df['VR_DESPESA_CONTRATADA'].astype(float)
```

```python
# Queremos trabalhar somente com Vereador

# Excluindo Registros que contém 'VICE-PREFEITO' da variável 'Cargo'

df.drop(df.loc[df['DS_CARGO']=='Vice-prefeito'].index, inplace=True)
df[['DS_CARGO','VR_DESPESA_CONTRATADA']].groupby(['DS_CARGO'], as_index=False).sum().sort_values(by='VR_DESPESA_CONTRATADA', ascending=False)

df.drop(df.loc[df['DS_CARGO']=='Prefeito'].index, inplace=True)
df[['DS_CARGO','VR_DESPESA_CONTRATADA']].groupby(['DS_CARGO'], as_index=False).sum().sort_values(by='VR_DESPESA_CONTRATADA', ascending=False)
```

```python
# Extraindo colunas que serão usadas

df2 = df[['NM_CANDIDATO','VR_DESPESA_CONTRATADA']].groupby(['NM_CANDIDATO'], as_index=False).sum().sort_values(by='VR_DESPESA_CONTRATADA', ascending=False)
df2
```

```python
# Renomeando as colunas

df2.columns = ['Nome_Completo', 'Valor_Total_Declarado']
df2
```

```python
# Salvando dataframe em um arquivo csv

df2.to_csv("dados_valor.csv", index=False)
```

```python
# Excluindo possíveis duplicatas

dataset = dataset.sort_values('Nome_Completo', ascending=False)
dataset = dataset.drop_duplicates(subset='Nome_Completo', keep='first')
dataset
```

```python
# Excluindo variáveis que não serão usadas

dataset = dataset.drop('Numero_Candidato_x', axis=1)
dataset = dataset.drop ('Nome_Completo', axis=1)
dataset = dataset.drop ('Nome_Urna', axis=1)
dataset = dataset.drop ('Numero_Candidato_y', axis=1)
dataset = dataset.drop ('Despesas_Max_Partido', axis=1)
dataset = dataset.drop ('Mun_Nascimento', axis=1)
dataset = dataset.drop ('Ocupacao', axis=1)
dataset = dataset.drop ('Votos', axis=1) # Por não sabermos o voto até o fim da eleição
dataset
```
